```python

def hhmmss_to_seconds(time_str):
    """
    Convert a time string in hh:mm:ss (or mm:ss) format to seconds.
    If already numeric, returns it as a float.
    """
    try:
        time_str = time_str.strip()
        parts = time_str.split(":")
        if len(parts) == 3:
            h, m, s = parts
            return float(h) * 3600 + float(m) * 60 + float(s)
        elif len(parts) == 2:
            m, s = parts
            return float(m) * 60 + float(s)
        else:
            return float(time_str)
    except Exception as e:
        print(f"Error converting '{time_str}' to seconds: {e}")
        return None

def parse_and_convert_snippet(snippet):
    """
    Parse a snippet entry (which might be a string like "[['00:00:34','00:01:34']]" or "Missing value")
    and convert the intervals to seconds. Supports either a single interval [start,end] or multiple intervals.
    Returns a list of intervals in seconds or None if invalid.
    """
    # 1) Handle the literal string "Missing value"
    if isinstance(snippet, str):
        snippet = snippet.strip()
        if snippet == "Missing value":
            return None
        
        # 2) If it's not "Missing value", try to parse it as a Python literal.
        try:
            snippet = ast.literal_eval(snippet)
        except Exception as e:
            print(f"Error parsing snippet: {snippet}. Error: {e}")
            return None
    
    # 3) If snippet is still not a list, return None
    if not isinstance(snippet, list):
        return None
    
    # 4) Check if it's a list of intervals (multiple) or a single interval
    def convert_interval_to_seconds(interval):
        if isinstance(interval, list) and len(interval) == 2:
            start = hhmmss_to_seconds(interval[0])
            end   = hhmmss_to_seconds(interval[1])
            return [start, end]
        return None
    
    # If snippet’s first element is itself a list => multiple intervals
    if snippet and isinstance(snippet[0], list):
        converted = []
        for inter in snippet:
            conv = convert_interval_to_seconds(inter)
            if conv is not None:
                converted.append(conv)
        return converted
    else:
        # Single interval e.g. ['00:00:34','00:01:34']
        conv = convert_interval_to_seconds(snippet)
        return [conv] if conv else None

# ----------------- Main Logic -----------------
# Suppose annot_data is your DataFrame with "hate_snippet" column
# We'll create a new column called "converted_hate_snippet"

annot_data["converted_hate_snippet"] = annot_data["hate_snippet"].apply(parse_and_convert_snippet)

# Display a few rows to verify
print(annot_data[["hate_snippet", "converted_hate_snippet"]].head(10))
```

```python
import ast
import pandas as pd

# 1) Build lookup maps from annot_data:
#    - intervals_map[video_name] -> list of (start, end) floats
#    - audio_map    [video_name] -> int(Audio_hate)
intervals_map, audio_map = {}, {}
for _, row in annot_data.iterrows():
    vid = row['video_name']
    raw = row.get('converted_hate_snippet')

    # Parse raw into a Python list of [start, end] pairs
    if isinstance(raw, list):
        intervals = raw
    elif isinstance(raw, str):
        try:
            intervals = ast.literal_eval(raw)
        except (ValueError, SyntaxError):
            intervals = []
    else:
        intervals = []

    # Coerce to floats
    intervals = [(float(s), float(e)) for s, e in intervals]
    intervals_map[vid] = intervals

    # Grab the audio flag (default to 0)
    audio_map[vid] = int(row.get('Audio_hate', 0))

# 2) Overlap helper: returns True if [start,end] overlaps any hate interval
def overlaps(start: float, end: float, hate_intervals: list[tuple[float,float]]) -> bool:
    for a, b in hate_intervals:
        if end >= a and start <= b:
            return True
    return False

# 3) Label each chunk in new_df
def label_chunk(row):
    vid = row['video']
    # If audio model never flagged hate for this video, it's 0
    if audio_map.get(vid, 0) == 0:
        return 0
    # Otherwise check for an overlap with any hate interval
    return int(overlaps(row['start'], row['end'], intervals_map.get(vid, [])))

new_df['Chunk_hate'] = new_df.apply(label_chunk, axis=1)

# 4) Now pull in the video‐level Hate_Label from annot_data
#    (so each chunk row also carries its video's overall hate label)
label_map = annot_data.set_index('video_name')['Hate_Label']
new_df['Hate_Label'] = new_df['video'].map(label_map)

# 5) (Optional) Save or inspect
new_df.to_csv("chunks_with_hate_labels.csv", index=False)
print(new_df[['video','start','end','Chunk_hate','Hate_Label']].head())
```

```python
def fill_missing_starts(df: pd.DataFrame) -> pd.DataFrame:
    """
    Detect missing start times, and:
      - for each video group, replace a NaN start with the previous row's end
      - if it's the first row in a video (so there's no previous end), fill with 0
    
    Assumes df has columns: ['video', 'start', 'end', …]
    """
    df = df.copy()
    # shift the end-times down 1 per video
    prev_ends = df.groupby('video')['end'].shift(1)
    
    # wherever start is NaN, grab the previous end; then any still-NaN (first in group) → 0
    df['start'] = (
        df['start']
          .fillna(prev_ends)
          .fillna(0)
    )
    
    return df
```

```python
def flag_and_fix_first_full(df: pd.DataFrame) -> pd.DataFrame:
    """
    1. Add a 'problem' column that's 1 when end < start, else 0.
    2. Identify rows that are both problematic and the first segment of their video,
       and set their start to 0.
    3. Return the full DataFrame (same size) with the updates.
    """
    df = df.copy()

    # Flag problematic rows
    df['problem'] = (df['end'] < df['start']).astype(int)

    # Identify first row in each video
    is_first = df.groupby('video').cumcount() == 0

    # For problematic first rows, reset start to 0
    mask = (df['problem'] == 1) & is_first
    df.loc[mask, 'start'] = 0

    return df

result = flag_and_fix_first_full(final_df)
#Recalculating the problem column
result["problem"] = (result['end'] < result['start']).astype(int)

#Seeing the problematic rows
print(result[result["problem"] == 1])
```

```python
# 2) In frames_data, rename 'video' → 'frame_file'
#    (because it actually contains filenames like 'hate_video_1_frame0.jpg')
frames_data = frames_data.rename(columns={'video':'frame_file'})

# 3) In ocr_data, build the exact same filename string
#    a) Extract the suffix from the existing frame column:
#       'video_1_frame240.jpg' → '_frame240.jpg'
ocr_data['suffix'] = ocr_data['frame'].str.extract(r'^video_\d+(_.*)$')[0]

#    b) Prepend the video id ('hate_video_1') to that suffix:
#       → 'hate_video_1_frame240.jpg'
ocr_data['frame_file'] = ocr_data['video'] + ocr_data['suffix']

# 4) Now merge on 'frame_file'
merged_ocr_image = pd.merge(
    frames_data, 
    ocr_data[['frame_file','text']],   # only need the text column from OCR
    on='frame_file',
    how='left'
).dropna()

# merged now has columns:
#   timestamp_seconds, frame_number, frame_file, text
#
# and `text` is the OCR result for each frame (or NaN if OCR data was missing).
print(merged_ocr_image.isna().sum())
# Example:
merged_ocr_image
```

```python
# --- 3) Extract video_name from frame_file ---
merged_ocr_image['video_name'] = (
    merged_ocr_image['frame_file']
    .str.split('_frame')
    .str[0]
)

# --- 4) Parse annot_data into a video_name -> [(start,end),…] map ---
intervals_map = {}
for _, row in annot_data.iterrows():
    vid = row['video_name']
    raw = row.get('converted_hate_snippet', None)

    # parse string/list into Python list of pairs
    if isinstance(raw, list):
        intervals = raw
    elif isinstance(raw, str):
        try:
            intervals = ast.literal_eval(raw)
        except (ValueError, SyntaxError):
            intervals = []
    else:
        intervals = []

    # coerce to floats and store
    clean = []
    for s,e in intervals:
        try:
            clean.append((float(s), float(e)))
        except:
            pass
    intervals_map[vid] = clean

# --- 5) Helper to check if a timestamp is in any interval ---
def in_hate(ts: float, ivals: list[tuple[float,float]]) -> bool:
    for a,b in ivals:
        if a <= ts <= b:
            return True
    return False

# --- 6) Label each frame row with Frame_hate ---
def label_frame(row):
    vid  = row['video_name']
    ts   = row['timestamp_seconds']
    iv   = intervals_map.get(vid, [])
    return int(in_hate(ts, iv))

merged_ocr_image['Frame_hate'] = merged_ocr_image.apply(label_frame, axis=1)

# --- merged_ocr_image is now ready ---
# It has columns:
#   frame_file, timestamp_seconds, frame_number, ocr_text,
#   video_name, Frame_hate

# Example inspection:
print("-------------------Merged_ocr_image_info-------------------")
print(merged_ocr_image['Frame_hate'].value_counts())


filtered_ocr_image = merged_ocr_image[
    merged_ocr_image['video_name'].isin(valid_videos)
].reset_index(drop=True)

# Example inspection:
print("-------------------Merged_ocr_image_filtered_info-------------------")
print(filtered_ocr_image['Frame_hate'].value_counts())

filtered_ocr_image.to_csv("./final_output_labelled/labelled_chunk_image_ocr_comb.csv")
```

```python
# 2) Prepare the frames table
frames = (
    frames_data
    # rename to the actual filename
    .rename(columns={'video':'frame_file'})
    # extract the video_id from that filename
    .assign(video_id=lambda df: df['frame_file'].str.split('_frame').str[0])
)

# 3) Prepare the OCR table and build the matching frame_file key
ocr = ocr_data.copy()
# extract suffix "_frameXXX.jpg" from the 'frame' column
ocr['suffix'] = ocr['frame'].str.extract(r'^video_\d+(_.*)$')[0]
ocr['frame_file'] = ocr['video'] + ocr['suffix']
# keep only the text we need
ocr = ocr[['frame_file', 'text']].rename(columns={'text':'ocr_text'})

# 4) Merge OCR text into frames
frames = pd.merge(frames, ocr, on='frame_file', how='left')
# now frames has: frame_file, timestamp_seconds, frame_number, video_id, ocr_text

# 5) Build a lookup by video_id
frames_by_vid = { vid:grp for vid, grp in frames.groupby('video_id') }

# 6) Iterate over your chunks and collect filenames + OCR strings
final = final_chunk_df.copy()
final['frames']    = None  # will hold lists of filenames
final['ocr_texts'] = None  # will hold lists of OCR strings

for idx, row in final.iterrows():
    vid   = row['video']   # e.g. 'hate_video_30'
    start = row['start']
    end   = row['end']
    
    # grab only the frames for this video (or an empty DF)
    grp = frames_by_vid.get(vid, pd.DataFrame(columns=frames.columns))
    
    # filter to the timestamp window
    mask = (grp['timestamp_seconds'] >= start) & (grp['timestamp_seconds'] <= end)
    sel  = grp.loc[mask]
    
    # store two parallel lists
    final.at[idx, 'frames']    = sel['frame_file'].tolist()
    final.at[idx, 'ocr_texts'] = sel['ocr_text'].tolist()

# 7) (Optional) overwrite your original
final_chunk_df = final

# Now each row of final_chunk_df has:
#   • frames   → List[str] of all matching frame filenames  
#   • ocr_texts → List[str] of all corresponding OCR’d strings
```

```python
# 1) Prepare frames_df
frames = (
    frames_data
    .rename(columns={'video':'frame_file'})        # avoid clashing with chunk_df 'video'
    .assign(
        video_id=lambda d: d['frame_file']
                         .str.split('_frame')       # 'hate_video_30_frame540.jpg' → ['hate_video_30','540.jpg']
                         .str[0]
    )
    # now frames has columns: frame_file, timestamp_seconds, frame_number, video_id
)

# 2) Group frames by video_id for fast lookup
frames_by_vid = dict(tuple(frames.groupby('video_id')))

# 3) Prepare your chunk DataFrame (no need to reset_index here)
final_chunk_df = final_chunk_df.copy()
final_chunk_df['frames'] = None  # placeholder

# 4) For each chunk, pull the matching frames
for idx, row in final_chunk_df.iterrows():
    vid   = row['video']          # e.g. 'hate_video_30'
    start = row['start']
    end   = row['end']
    
    # get only the frames for this video (or empty DataFrame)
    sub = frames_by_vid.get(vid, pd.DataFrame(columns=frames.columns))
    
    # filter by timestamp
    mask = sub['timestamp_seconds'].between(start, end, inclusive='both')
    matched = sub.loc[mask, 'frame_file'].tolist()
    
    final_chunk_df.at[idx, 'frames'] = matched

# Done! final_chunk_df['frames'] now contains exactly the list of 
# frame filenames from the correct video whose timestamps lie in [start, end].
```
