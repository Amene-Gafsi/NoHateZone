```python
# !pip install python-doctr[torch]
from doctr.io import DocumentFile
from doctr.models import ocr_predictor

model = ocr_predictor(pretrained=True)
```

```python
doc = DocumentFile.from_images("../../data/detect2.jpeg")

result = model(doc)

print(result.render())  # the detected text in the image
print(result.export()['pages'][0]['blocks'])  # We have the exact text position in the image and the probability of the prediction
```
